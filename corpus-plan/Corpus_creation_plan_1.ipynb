Импортируем библиотеки и модули, необходимые для работы с текстом. Помимо `pymorphy2` используем также инструменты библиотеки `nltk` (Natural Language Toolkit).

````python
import pymorphy2
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt')
````

Откроем и прочитаем файл .txt. Сегментируем текст на слова/токены и предложения. Результаты сохраним в виде списков в переменных `tokenized_text` и `sent_text`

````python
file = open('drive/MyDrive/Krongauz.txt','r',encoding='utf-8')
text = file.read()
string = text.replace("\n"," ")
tokenized_text = word_tokenize(text)
sent_text = sent_tokenize(text)
````

Назначим переменную `morph` для морфологического анализатора библиотеки `pymorphy2`.

````python
morph = pymorphy2.MorphAnalyzer()
````

### Многоуровневая разметка

Создадим пустой список `alignments_list`, в который будет последовательно добавляться разметка для каждого элемента списка списков `tokenized_sent`.

Создаем пустой двухуровневый список (список списков) с количеством вложенных списков, равным количеству предложений в списке `sent_text`

````python
alignments_list = [[] for x in range(len(sent_text))]
#проверяем длину списка
print(len(alignments_list))
````

Далее создадим многоуровневый список разметок, в котором будет три индекса: уровень предложений, уровень списка возможных разметок и конкретная разметка.

````python
for i in range(len(tokenized_sent)):
  for j in range(len(tokenized_sent[i])):
    alignments_list[i].append(morph.parse(tokenized_sent[i][j]))
````

Проверим, что список разметок устроен нужным образом

Список всех разметок для первого предложения.

````python
alignments_list[0]
````

Список всех разметок для пятого слова первого предложения.

````python
alignments_list[0][4]
````

Первая разметка для пятого слова первого предложения.

````python
alignments_list[0][4][0]
````

Отметим, что при наличии нескольких вариантов разметки `pymorphy2` выдает все возможные варианты в порядке убывания вероятности. Отдельно создадим список наиболее вероятных вариантов разметки (по тому же принципу, что и ранее: многоуровневый список).

````python
probable_alignment = [[] for x in range(len(alignments_list))]
#проверяем длину списка
print(len(probable_alignment))
````

````python
for i in range(len(alignments_list)):
  for j in range(len(alignments_list[i])):
    probable_alignment[i].append(alignments_list[i][j][0])
````

Вероятная разметка для слов первого предложения

````python
probable_alignment[0]
````

Из наиболее вероятного варианта разметки получим лемму (начальную форму) и теги.

````python
tags = [[] for x in range(len(probable_alignment))]
lemmas = [[] for x in range(len(probable_alignment))]
for i in range(len(probable_alignment)):
  for j in range(len(probable_alignment[i])):
    tags[i].append(probable_alignment[i][j].tag)
    lemmas[i].append(probable_alignment[i][j].normal_form)
````

````python
lemmas[0]
````

Можно представить списки списков в виде таблицы (а еще они имеют синхронизированные индексы, по которым можно получить нужную информацию для каждого элемента в многоуровневых списках)

````python
text_data_per_sent = pd.DataFrame(
    data={
        'предложение': sent_text,
        'словоформа' : tokenized_sent,
        'варианты разметки': alignments_list,
        'вероятная разметка': probable_alignment,
        'лемма' : lemmas,
        'тэги': tags
    })
````

````python
text_data_per_sent.head()
````

### Сплошная разметка

````python
sent_list_for_index = []
for i in range(len(tokenized_sent)):
  for j in range(len(tokenized_sent[i])):
    sent_list_for_index.append(sent_text[i])
````

````python
len(sent_list_for_index)
````

````python
alignments_list_full = []
for word in tokenized_text:
  alignments_list_full.append(morph.parse(word))
````

````python
len(alignments_list_full)
````

````python
probable_alignment_full = [i[0] for i in alignments_list_full]
len(probable_alignment_full)
````

````python
tags_full = []
lemmas_full = []
for alignment in probable_alignment_full:
    tags_full.append(alignment.tag)
    lemmas_full.append(alignment.normal_form)
len(tags_full)
````

Создадим таблицу "токен"-"варианты разметки"-"вероятная разметка"-"лемма"–"теги"

````python
text_data = pd.DataFrame(
    data={
        'предложение': sent_list_for_index,
        'словоформа' : tokenized_text,
        'варианты разметки': alignments_list_full,
        'вероятная разметка': probable_alignment_full,
        'лемма' : lemmas_full,
        'тэги': tags_full
    }, index=[sent_list_for_index, tokenized_text])
````

### Ручная разметка

Предлагается создание списка, содержащего морфологичекую разметку каждой словоформы, при этом в случае неединственного варианта разметки пользователю предлагается выбрать один из вариантов по индексу.

````python
alignments_list_full[0]
````

Создадим пустой список `manual_alignment`, в который последовательно будем добавлять разметку для каждой словоформы из списка `tokenized_text`. При этом в случае если вариант разметки только один, он добавляется в список автоматически, а в случае если вариантов больше одного, нужно выбрать нужный вариант из предложенных по индексу. Для удобства выводится сама словоформа и предложение, в котором она употреблена.

**NB! Важно помнить, что индексация начинается с 0, то есть первый вариант (часто он же наиболее вероятный) будет иметь индекс 0.**

````python
manual_alignment = []
for i in range(len(alignments_list_full[0:20])): # пока ограничиваемся несколькими первыми словоформами для тестирования
    if len(alignments_list_full[i]) == 1:
        manual_alignment.append(alignments_list_full[i])
    if len(alignments_list_full[i]) > 1:
        print(f"Словоформа: {tokenized_text[i]}")
        print(f"Предложение: {sent_list_for_index[i]}")
        for j in range(len(alignments_list_full[i])):
            print(j, alignments_list_full[i][j])
        manual_alignment.append(alignments_list_full[i][int(input(f"Введите индекс "))])
````
